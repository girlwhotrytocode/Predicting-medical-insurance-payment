2. Conduct ANOVA test (Age range, Sex, BMI range, Childern categorical, Smoker, Region --- Charges)  

```python
from scipy.stats import f_oneway
from statsmodels.formula.api import ols
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

* ANOVA test between Age range and Charges

```python
insurance.boxplot('charges', by='age_group', figsize=(12, 8))
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
aov = ols('charges ~ age_group', data = insurance).fit()
aov_table = sm.stats.anova_lm(aov, typ = 2 )
aov_table
```

```python
pair_t = aov.t_test_pairwise('age_group')
pair_t.result_frame
```

```python
tukey = pairwise_tukeyhsd(endog=insurance['charges'],
                          groups=insurance['age_group'],
                          alpha=0.05)
print(tukey)
```

It show that there are **strong** significant difference between the levels of the age group since the p-value is 0 (<0.05)  

Age 10s-30s, 40s-50s, over 60 are in the different group.

## Standard Linear Regression 

1. generating pipeline 

```python
features = ['age', 'sex', 'bmi', 'children', 'smoker', 'region']

X = insurance[features]
```

```python
y = insurance['charges']

y
```

```python
#split data 75/25, random state = 0 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=.25,
                                                    random_state = 0
                                                   )
np.sum(y_test)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
#Machine learning pipeline

#Age, BMI, Childern - std 
#sex, smoker region - onehot encoding & dummy variable
```

```python
#sex, smoker region
imp_ohe = make_pipeline(SimpleImputer(strategy='most_frequent'), 
                       OneHotEncoder(sparse=False))
#Age, BMI, Childern
imp_std = make_pipeline(SimpleImputer(), 
                       StandardScaler())
```

```python
preprocessor = make_column_transformer((imp_ohe, ['sex', 'smoker', 'region']), 
                       (imp_std, ['age', 'bmi', 'children']), 
                       remainder = 'passthrough')
```

```python
np.set_printoptions(edgeitems = 6, suppress = True) 
# onehot encoding ('sex', 'smoker', 'region') /standardization (age', 'bmi', 'children') 

preprocessor.fit_transform(X_train).round(3)
```

```python
preprocessor.fit_transform(X_train).round(3)
```

```python
#Standard linear regression 
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
pipe = make_pipeline(preprocessor, reg) 
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python

#R-square
from sklearn.metrics import r2_score
r2_score(y_test, y_pred).round(3)

```

```python
import math
```

```python
#MSE
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
math.sqrt(mse)
```

```python
#MAPE 
(np.mean(np.abs((y_test - y_pred)/y_test))*100)
```

```python
#PR 
sum(y_pred)/sum(y_test)
```

#### Performance when alpha, lambda changes and Cross-Validation = 10
to figure out the appropriate lambda value

```python
#selecting the best model 
from sklearn.model_selection import GridSearchCV 
```

```python
pipe.get_params()
```

```python
param_grid = { 'elasticnet__alpha': np.arange(0, 1, 0.1),
              'elasticnet__l1_ratio': np.arange(0, 1, 0.01)
}
```

```python
from sklearn.model_selection import KFold

cv = KFold(n_splits=10, shuffle=True, random_state = 0)
```

```python
#RMSE
gs = GridSearchCV(pipe, param_grid = param_grid, cv = cv, scoring='neg_root_mean_squared_error', n_jobs=-1)
gs.fit(X_train, y_train)
print(gs.best_params_)
print(gs.best_score_)
```

```python
#R2
gs = GridSearchCV(pipe, param_grid = param_grid, cv = cv, scoring='r2', n_jobs=-1)
gs.fit(X_train, y_train)
print(gs.best_params_)
print(gs.best_score_)
```

Decide to use 0.5 of lambda for a fair balance between R-squared and RMSE

#### Performance when alpha changes and Cross-Validation = 10

```python
from sklearn.model_selection import KFold

cv = KFold(n_splits=10, shuffle=True, random_state = 0)
```

```python
#Change alpha from 0 to 1 (0, 0.1, 0.2, 0.3, ... 1.0)
from sklearn.model_selection import cross_val_score
```

```python
for k in np.arange(0.1, 1.01, 0.1): 
    Elsnet = ElasticNet(alpha = k,l1_ratio = 0.5 ,  random_state = 0)
    pipe = make_pipeline(preprocessor, Elsnet) 
    score = cross_val_score(pipe, X, y, cv=cv, scoring='r2').mean().round(3)
    
    print('k=',k.round(3), ';mean R-squared:', score.mean().round(3))
```

```python
for k in np.arange(0.1, 1.01, 0.1): 
    Elsnet = ElasticNet(alpha = k, random_state = 0)
    pipe = make_pipeline(preprocessor, Elsnet) 
    score = cross_val_score(pipe, X, y, cv=cv, scoring='neg_root_mean_squared_error').mean().round(3)
    
    print('k=',k.round(3), ';mean RMSE:', score.mean().round(3))
```

```python
#MAPE
from sklearn.model_selection import cross_val_predict

for k in np.arange(0.1, 1.01, 0.1): 
    Elsnet = ElasticNet(alpha = k, random_state = 0)
    pipe = make_pipeline(preprocessor, Elsnet) 
    y_pred = cross_val_predict(pipe, X, y, cv=cv)

    MAPE = (np.mean(np.abs((y - y_pred)/y_test))*100).round(3)
    #print(y_pred)
    #print(y)
    print('k=',k.round(3), ';MAPE:', MAPE.round(3))
```

```python
#PR 

for k in np.arange(0.1, 1.01, 0.1): 
    Elsnet = ElasticNet(alpha = k, random_state = 0)
    pipe = make_pipeline(preprocessor, Elsnet) 
    y_pred = cross_val_predict(pipe, X, y, cv=cv)
    PR = (sum(y_pred)/sum(y)).round(3)
  
    print('k=',k.round(3), ';PR:', PR.round(3))
```
